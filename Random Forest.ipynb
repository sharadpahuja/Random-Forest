```python
from sklearn.preprocessing import LabelEncoder
lb=LabelEncoder()
df['Attrition']=lb.fit_transform(df['Attrition'])
```

```python
df_num = df[df.select_dtypes(include=np.number).columns]
df_cat = df[df.select_dtypes(include='object').columns]
df_dummy= pd.get_dummies(df_cat,drop_first=True)# drop first equal true
data =pd.concat([df_num,df_dummy],axis=1)
data.head()
```

# Random Forest

```python
rf_model = ensemble.RandomForestClassifier(n_estimators=150,criterion='entropy',
                                          class_weight='balanced',
                                           max_leaf_nodes=12,
                                          random_state=15)
rf_model.fit(x_train, y_train)

y_pred_test = dt_model.predict(x_test)
y_pred_train = dt_model.predict(x_train)

print('Accuracy Train:',metrics.accuracy_score(y_train,y_pred_train))
print('Accuracy Test:',metrics.accuracy_score(y_test,y_pred_test))
# Accuracy of train nd test should be close and should propotionate increase together
# If accuracy of train is more than 5-6 percent than test then we have to check if there is a variance error(overfitting)
# If accuracy of test is more than train data it means it has bias error (underfitting)
```

# Optimising Techniques to tune Hyper Parameters

# Random Search CV

```python
model_params = {'n_estimators' : [140,145,150,155,160], 'max_leaf_nodes' : range(10,20), 'criterion' : ['gini', 'entropy']}
```

```python
rf_model = ensemble.RandomForestClassifier(random_state = 15)
random_search_object = model_selection.RandomizedSearchCV(rf_model, model_params, n_iter = 10, cv = 3, random_state= 15)
# In random_search_best_model - It creates the model on the basis of  best (specified n_iter) only
```

```python
random_search_best_model = random_search_object.fit(x_train, y_train)
```

```python
y_pred_test = random_search_best_model.predict(x_test)
y_pred_train = random_search_best_model.predict(x_train)


print('Accuracy Train:',metrics.accuracy_score(y_train,y_pred_train))
print('Accuracy Test:',metrics.accuracy_score(y_test,y_pred_test))
```

```python
random_search_best_model.best_params_
```

# Grid Search

```python
grid_search_object = model_selection.GridSearchCV(rf_model, model_params, cv = 3)
# grid_search_object builds model on the basis of every possible outcome and chooses the best one to build a model among it
```

```python
grid_search_best_model = grid_search_object.fit(x_train, y_train)
```

```python
y_pred_test = grid_search_best_model.predict(x_test)
y_pred_train = grid_search_best_model.predict(x_train)


print('Accuracy Train:',metrics.accuracy_score(y_train,y_pred_train))
print('Accuracy Test:',metrics.accuracy_score(y_test,y_pred_test))
```
